```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset
dfn_finaciero_=pd.read_csv('/content/drive/MyDrive/Analisis_Datos_Basico/Seccion 20/datos_financieros.csv')

# Seleccionar las columnas relevantes para el análisis
columns_of_interest = ['Edad', 'Genero', 'Historial_Crediticio', 'Deuda']
df_selected = dfn_finaciero_[columns_of_interest].dropna()  # Eliminamos filas con valores nulos

# Convertir la columna 'Historial_Crediticio' a categórica (si es necesario)
df_selected['Historial_Crediticio'] = df_selected['Historial_Crediticio'].astype(str)

# --- Pregunta 1: ¿Las personas con buen historial tienen menor deuda? ---
plt.figure(figsize=(8, 6))
sns.boxplot(x='Historial_Crediticio', y='Deuda', data=df_selected, palette="Set2")
plt.title('Relación entre Historial Crediticio y Deuda', fontsize=14, color='purple')
plt.xlabel('Historial Crediticio', fontsize=12)
plt.ylabel('Deuda', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# --- Pregunta 2: ¿Cómo varía el historial crediticio según edad y género? ---
plt.figure(figsize=(10, 8))
sns.boxplot(x='Genero', y='Edad', hue='Historial_Crediticio', data=df_selected, palette="coolwarm")
plt.title('Variación del Historial Crediticio por Edad y Género', fontsize=14, color='green')
plt.xlabel('Género', fontsize=12)
plt.ylabel('Edad', fontsize=12)
plt.legend(title='Historial Crediticio', loc='upper right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

#Gráfico 1: Relación entre Historial Crediticio y Deuda
#Observación:

#Los clientes con un historial "Malo" tienden a tener mayores niveles de deuda en comparación con aquellos con un historial "Bueno" o "Regular".
#Los clientes con un historial "Bueno" muestran menos variabilidad en sus niveles de deuda.
#Conclusión:

print("Las personas con un historial crediticio más favorable suelen manejar niveles más bajos y consistentes de deuda, lo que refuerza su fiabilidad financiera")
```
