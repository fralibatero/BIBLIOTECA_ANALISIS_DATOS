1.	Comparación de calificaciones de estudiantes en diferentes asignaturas
Se quiere comparar las calificaciones promedio de los estudiantes en distintas materias para detectar diferencias de rendimiento.
•	asignaturas =Matemáticas,Ciencias,Historia,Inglés,Arte
•	calificaciones = 85, 90, 78, 88, 92


```python
import pandas as pd
import matplotlib.pyplot as plt

datos = {'asignaturas': ['Matemáticas', 'Ciencias', 'Historia', 'Inglés', 'Arte'],
         'calificaciones': [85, 90, 78, 88, 92]}

df = pd.DataFrame(datos)

plt.bar(df['asignaturas'], df['calificaciones'], color=['magenta', 'green', 'red', 'orange', 'purple'])
plt.xlabel('Asignaturas')
plt.ylabel('Calificaciones')
plt.title('Calificaciones de Estudiantes')
plt.show()
plt.tight_layout()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

horas_estudio = [5,10,15,20,25,30,35,40,45,50],
Calificaciones = [60,65,70,75,80,85,88,90,92,95],

plt.scatter(horas_estudio, Calificaciones, color='blue', s=100, edgecolors= 'black')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificaciones')
plt.title('Relación entre Horas de Estudio y Calificaciones')
```


**4.	Distribución de salarios en tres departamentos distintos
Se tiene un conjunto de datos de salarios en diferentes departamentos y quieres visualizar la variabilidad y mediana de los salarios en cada uno.**

datos_salarios = {
'Departamento': ['Recurso Humanos]*10 + ['IT']*10 + [administración]*10,
'Salario': [3000, 3200, 3400, 3600, 3800, 4000, 4200, 4400, 4600, 4800,
5000, 5200, 5400, 5600, 5800, 6000, 6200, 6400, 6600, 6800,
7200, 7400, 7600, 7800, 8000, 8200, 8400, 8600, 8800, 9000]
}


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

datos_salarios = {'Departamento': ['Recurso Humanos'] * 10 + ['IT'] * 10 + ['Administración'] * 10,
                  'Salario': [3000, 3200, 3400, 3600, 3800, 4000, 4200, 4400, 4600, 4800,
                             5000, 5200, 5400, 5600, 5800, 6000, 6200, 6400, 6600, 6800,
                             7200, 7400, 7600, 7800, 8000, 8200, 8400, 8600, 8800, 9000]}

df = pd.DataFrame(datos_salarios)

plt.figure(figsize=(10, 6))
sns.boxplot(x='Departamento', y='Salario', data=df, hue='Departamento', palette='pastel', legend=False)
plt.xlabel('Departamento', fontsize=16)
plt.ylabel('Salario', fontsize=16)  # Increased font size for ylabel
plt.title('Distribución de Salarios por Departamento', fontsize=18)  # Added title
plt.tight_layout()  # Prevents overlapping labels
plt.show()
```
